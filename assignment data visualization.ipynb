# seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
print(sns.get_dataset_names())
```

```python
df=sns.load_dataset("diamonds")
```

```python
df.head(5)
```

```python
df.info()
```

```python
sns.relplot(data=df,x="carat",y="table",hue="color")
```

#distplot

```python
sns.distplot(df.price,kde=True)
```

```python
sns.distplot(df.price,kde=False)
```

```python
df.head()
```

```python
sns.barplot(data=df,x="cut",y="table",hue="color")
```

# count plot

```python
sns.countplot(data=df,x="cut")
```

# joint plot

```python
sns.jointplot(data=df,x="carat",y="price")
```

# boxplots

```python
sns.boxplot(x="color",y="carat",data=df)
```

```python
#correlation 

df
```

```python
corr=df.corr()
```

```python
sns.heatmap(corr,annot=True,cmap="YlGnBu")
```

# pairplot

```python
sns.pairplot(data=df,hue="cut",kind="scatter",palette="husl")
```
